```python
import pandas as pd
import html
import re
from xml.etree.ElementTree import Element, SubElement, tostring
from xml.dom.minidom import parseString

def safe_text(value):
    """Convert to string, escape XML-invalid chars, strip control chars."""
    if pd.isna(value):
        return ""
    text = str(value)
    # Remove control characters that XML parsers reject
    text = re.sub(r"[\x00-\x08\x0B\x0C\x0E-\x1F]", "", text)
    return html.escape(text)

def df_to_kml(df, output_path):
    kml = Element('kml', xmlns='http://www.opengis.net/kml/2.2')
    document = SubElement(kml, 'Document')

    for idx, row in df.iterrows():
        try:
            lon = row.get('longitude')
            lat = row.get('latitude')
            if pd.isna(lon) or pd.isna(lat):
                continue
            lon, lat = float(lon), float(lat)

            placemark = SubElement(document, 'Placemark')

            name = SubElement(placemark, 'name')
            name.text = safe_text(row.get('Tipo de evento', 'Sin nombre'))

            description = SubElement(placemark, 'description')
            description.text = (
                f"Evento: {safe_text(row.get('Evento', ''))}\n"
                f"Fecha: {safe_text(row.get('Fecha del evento', ''))}\n"
                f"Departamento: {safe_text(row.get('departamento', ''))}\n"
                f"Municipio: {safe_text(row.get('municipio', ''))}\n"
                f"Presunto actor responsable: {safe_text(row.get('Presunto actor responsable', ''))}\n"
                f"Descripción: {safe_text(row.get('descripcion', ''))}\n"
                f"Estatus: {safe_text(row.get('estatus', ''))}\n"
                f"Cantidad de artefactos: {safe_text(row.get('cantidad de artefactos', ''))}"
            )

            point = SubElement(placemark, 'Point')
            coordinates = SubElement(point, 'coordinates')
            coordinates.text = f"{lon},{lat},0"

        except Exception as e:
            print(f"Skipping row {idx} due to error: {e}")
            continue

    # tostring returns bytes → decode to str
    xml_bytes = tostring(kml, encoding="utf-8")
    xml_str = xml_bytes.decode("utf-8")

    # Now parseString safely
    dom = parseString(xml_str)
    pretty_xml = dom.toprettyxml(indent="  ", encoding="UTF-8")

    with open(output_path, 'wb') as f:
        f.write(pretty_xml)
```

```python
import pandas as pd
import html
import re
import zipfile
from xml.etree.ElementTree import Element, SubElement, tostring
from xml.dom.minidom import parseString

def safe_text(value, maxlen=60):
    """Convert to string, escape XML-invalid chars, strip control chars, and truncate."""
    if pd.isna(value):
        return ""
    text = str(value)
    text = re.sub(r"[\x00-\x08\x0B\x0C\x0E-\x1F]", "", text)  # remove control chars
    text = html.escape(text)
    if len(text) > maxlen:
        text = text[:maxlen] + "..."
    return text

def df_to_kmz(df, output_path, n=200):
    """
    Convert a random sample of N rows from a DataFrame to KMZ.
    Default N=200.
    """
    # Random sample
    if len(df) > n:
        df = df.sample(n=n, random_state=42)  # reproducible sample
    else:
        df = df.copy()

    kml = Element('kml', xmlns='http://www.opengis.net/kml/2.2')
    document = SubElement(kml, 'Document')

    # Define a simple pushpin style once
    style = SubElement(document, 'Style', id="pushpin")
    icon_style = SubElement(style, 'IconStyle')
    icon = SubElement(icon_style, 'Icon')
    href = SubElement(icon, 'href')
    href.text = "http://maps.google.com/mapfiles/kml/pushpin/red-pushpin.png"

    for idx, row in df.iterrows():
        try:
            lon = row.get('longitude')
            lat = row.get('latitude')
            if pd.isna(lon) or pd.isna(lat):
                continue
            lon, lat = float(lon), float(lat)

            placemark = SubElement(document, 'Placemark')

            # Use integer index as name
            name = SubElement(placemark, 'name')
            name.text = str(idx)

            style_url = SubElement(placemark, 'styleUrl')
            style_url.text = "#pushpin"

            description = SubElement(placemark, 'description')
            description.text = (
                f"Tipo: {safe_text(row.get('Tipo de evento', ''))}\n"
                f"Evento: {safe_text(row.get('Evento', ''))}\n"
                f"Fecha: {safe_text(str(row.get('Fecha del evento', ''))[:10])}\n"
                f"Departamento: {safe_text(row.get('departamento', ''))}\n"
                f"Municipio: {safe_text(row.get('municipio', ''))}\n"
                f"Actor: {safe_text(row.get('Presunto actor responsable', ''))}\n"
                f"Descripción: {safe_text(row.get('descripcion', ''), maxlen=30)}\n"
                f"Estatus: {safe_text(row.get('estatus', ''))}"
            )

            point = SubElement(placemark, 'Point')
            coordinates = SubElement(point, 'coordinates')
            coordinates.text = f"{lon},{lat},0"

        except Exception as e:
            print(f"Skipping row {idx} due to error: {e}")
            continue

    # Build XML string
    xml_bytes = tostring(kml, encoding="utf-8")
    xml_str = xml_bytes.decode("utf-8")
    dom = parseString(xml_str)
    pretty_xml = dom.toprettyxml(indent="  ", encoding="UTF-8")

    # Write KMZ
    with zipfile.ZipFile(output_path, 'w', compression=zipfile.ZIP_DEFLATED) as kmz:
        kmz.writestr("doc.kml", pretty_xml)
```
